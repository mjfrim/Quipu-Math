```python
##Main calculation

#This large for-loop is to repeat the process 100,000 times
#First, we find the best permutation for a list of nine random numbers
#The best order is the one that minimizes the average percent errors
#Then we aggregate data and collect the percent errors of each of these 100,000 sets of nine numbers once they have been optimized
#The reason is it 100000-count is so that you can stop and start the program as it does take a while, ~10hrs

#You can do this much faster with asynchronous techniques
#But due to my lack of coding experience, I understood the steps going on here much better, so I stuck to this
#My attempt at an asyncrhonous code allowed for 10,000 iteration in ~20mins, so it would cut doen the time significantly

for i in range(100000-count):
    
    #Make a list of 9 random numbers
    rand_list=random.sample(range(500,3000),9)
    
    #Start with a number so large that it is broken immediately
    min_avg_temp=10
    
    #Goes through each permutation of the nine numbers
    for item in list(permutations(rand_list)):
        
        #Calculates the percent error for that perumation
        percents_iter=[abs(item[0]*item[4]-item[6]*item[7])/(item[0]*item[4]),
         abs(item[1]*item[5]-item[7]*item[8])/(item[1]*item[5]),
         abs(item[2]*item[3]-item[8]*item[6])/(item[2]*item[3])]
        
        #Calculates the mean of the three percent errors
        min_avg_iter=sum(percents_iter)/3
        
        #If that mean is smaller than the previous lowest mean, the relevant information is collected
        #Instead of creating a massive list of percent errors and averages,
        #This goes through the list and only remembers the smallest one
        #If there is a subsequent that is smaller, it forgets the previous info and replaces it
        #So it only remembers one set of percent errors, one average, and one order at a time
        #This is significantly faster than creating a list of all of them and then finding the minima later
        if min_avg_iter < min_avg_temp:
            min_avg_temp = min_avg_iter
            min_percents_temp = percents_iter
            sixlet_temp = item
        
    #At the end, once the optimized permutation is found, the percent errors, the average of them, and the ordered list itself are all added to the larger list of 100,000 items
    min_avgs.append(min_avg_temp)
    min_percents.append(min_percents_temp)
    sixlets.append(sixlet_temp)
    
    #Use this to keep track of your progress
    #print(count)
    count=count+1
```

```python
##Timecheck
#Same code as the previous cell, but this is set up to measure the time elapsed in calculation
#Designed to pick a number of iterations (e.g. 1 or 20) and measure the time it takes (about .6 seconds per iteration)
#so as to estimate the time it will take for the 100,000 iterations (>10hrs)

start = time.time()
n_iterations = 20

for i in range(n_iterations):
    rand_list=random.sample(range(500,3000),9)
    min_avg_time=10
    for item in permutations(rand_list):
        percents_iter=[abs(item[0]*item[4]-item[6]*item[7])/(item[0]*item[4]),
         abs(item[1]*item[5]-item[7]*item[8])/(item[1]*item[5]),
         abs(item[2]*item[3]-item[8]*item[6])/(item[2]*item[3])]
        min_avg_iter=sum(percents_iter)/3
        if min_avg_iter < min_avg_time:
            #These are not lists just because it was easier for the n=1 case, the number of actual steps is the same, and this is all just for estimation anyways
            min_avg_time = min_avg_iter
            min_percents_time = percents_iter
            sixlet_time = item
end = time.time()

print(end - start)
print(min_avg_time)
print(min_percents_time)
print(sixlet_time)
```
